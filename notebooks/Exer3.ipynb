```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler,LabelEncoder

df_iris = pd.DataFrame({
    'tailles_Cm': [150, 180, 160, 200, 500],
    'couleurs': ['rouge', 'vert', 'bleu', 'rouge', 'vert']
})
# Normalisation
scaler = MinMaxScaler()
df_iris['tailles_Cm'] = scaler.fit_transform(df_iris[['tailles_Cm']])

# Encodage
encoder = LabelEncoder()
df_iris['couleurs'] = encoder.fit_transform(df_iris['couleurs'])
```
